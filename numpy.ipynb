```python
plt.imshow(im[:,:,0],'Greens')
```

```python
im[:,:,0]
```

```python
n=im.max(axis=2)
n.shape
```

```python
plt.imshow(n,'gray')
```

```python
n=(im[:,:,0].astype('uint64')+im[:,:,1].astype('uint64')+im[:,:,2].astype('uint64'))/3
plt.imshow(n,'gray')
```

```python
n=im[:,:,0]//3+im[:,:,1]//3+im[:,:,2]//3
plt.imshow(n,'gray')
```

```python
n=im.mean(axis=2)
plt.imshow(n,'gray')
```

```python
m=np.array([0.2126,0.7152,0.0722])
n=im[:,:,0]*0.2126+im[:,:,1]*0.7152+im[:,:,2]*0.0722
plt.imshow(n,'gray',vmin=0,vmax=255)
```

```python
im2=im.dot(m)
plt.imshow(im2,'gray',vmin=0,vmax=255)
```
